Pig Latin is a common children’s “secret” language. The rules for translating words from English into Pig Latin are pretty straightforward:
<p></p>
<li>If the word begins with a vowel (a, e, i, o, or u), add “way” to the end of the word. <i>So “eye” becomes “eyeway” and “ant” becomes “antway.”</i></li>
<li>If the word begins with any other letter, then we take the first letter, put it at the end of the word, and then add “ay.”
<i> So “delta” becomes “eltaday” and “hoboken” becomes “obokenhay.”</i></li>

<p></p>
Given a list of english words, translate all words into Pig Latin using Python and save them in output list. Maintain the same order as the input words.

input_list = ["House", "Elephant", "Island"]
<p></p>
output_list = ['ousehay', 'elephantway', 'islandway']
