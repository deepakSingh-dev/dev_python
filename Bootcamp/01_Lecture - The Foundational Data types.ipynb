## Number
<font color=darkblue> We will cover Integers and Floats here. Additionally, Python supports complex number and other number types which we will not cover today.</font>

````python
# Integer
x = 3
# Float
y = 2.5
````

<font color=darkblue>
<font color=red> print()</font> will output to console and it can also evaluate expressions.
</font>

````python
print(x,y)
print()
print('x+y is:', x+y)
````

<font color=darkblue>
- Addition (<font color=red>+</font>), Subtraction (<font color=red>-</font>), Multiplication (<font color=red>*</font>) operators work as expected between numbers.<br>
- Parentheses <font color=red>()</font> can be used for grouping.
</font>

````python
print('x =',x,'y =',y)
print(x + y)
print(x - y)
print(x * y)
print(x + (x * y))
````

<font color=darkblue>
- Division (<font color=red>/</font>) always returns a float.<br>
- Floor division (<font color=red>//</font>) returns an integer discarding any fractional portion.<br>
- Remainder (<font color=red>%</font>) takes the sign of the denominator.<br>
- Power operator (<font color=red>**</font>) can be used to return powers.<br>
</font>

````python
# Division
print('Simple division:', 4/2, end='\n*****\n')

# Floor division
print('Floor division:',4//2, end='\n*****\n')
# With negative numbers, it will return the largest integer less than or equal to result of division
print('Floor division with -ve number:', 7//-2, end='\n*****\n')

# Remainder a.k.a. modulo operation
print('Modulo operator:', 7%2, end='\n*****\n')
print('Modulo operator with a -ve denominator:',7%-2 ,end='\n*****\n')
print('Modulo operator with a -ve numerator:', -7%2, end='\n*****\n')

# Power operation
print('Power operator:', 7**2 ,end='\n*****\n')
````

## String

<font color=darkblue>
- Strings need to be enclosed in quotes when they are created. Quotes can be single or double or triple, it does not matter. Just keep it consistent.<br>
- String manipulation is a strong suite of Python (we will cover in later lectures).<br><br>
</font>

````python
a = 'Welcome to Stevens!'
print(a)

# \ can be used to escape quotes
b = 'Ducks are Stevens\' athletes.'
print(b)

# or you can simply change the quotes level
c = "Ducks are Stevens' athletes."
print(c)
````

## Boolean
<font color=darkblue> A boolean can assume only two values - <font color=red>True</font> or <font color=red>False</font>.</font>

````python
x, y, z = 5, 10, 15

print('x == y:', x == y)  # check x is equal to y
print('x != y:', x != y)  # check x is not equal to y
print('x > y:', x > y)    # check x is greater than y
print('x < y:', x < y)    # check x is less than y

# Evaluate an expression and assign boolean value to a variable
z = x == y
print('z:', z)

# Directly assign the boolean value to a variable
i = True
print('i:', i)
j = False
print('j:', j)

# An empty boolean variable can be instantiated
z = bool()
print('Empty:', z)  # empty = False, something = True

z = bool(10)
print('Something:', z)
"""
# Key to remember:
For numbers, any non-zero value resolves to True; zero is false.
For a string or list or any sequence, anything with a non-zero length also resolves to True; empty sequences are False.
"""
print(bool(0))
print(bool(""))
````

## List
<font color=darkblue> Python's list data structure is one of its key distinguishing features and also one of the most versatile.</font>

````python
# A list can contain multiple data types
a = [10, 'a string can be in a list', 20, 30]
print('a: ', a)

# A list can also contain other variables
b = [1, 'x', x]
print('b: ', b)

# A list can also contain list of lists
c = [5, 10, 'Hoboken', b]
print('c: ', c)

# A list can be instantiated without any elements
d = [] # list()
print('d: ', d)
````

## Tuple
<font color=darkblue> A tuple is just like a list but it is immutable (i.e., cannot be changed) </font>

````python
# A tuple can contain multiple data types
a = (10, 'a string can be in a list', 20, 30)
print('a: ', a)

# A tuple can also contain other variables
b = (1, 'x', x)
print('b: ', b)

# A tuple can also contain list of lists
c = (5, 10, 'Hoboken', b)
print('c: ', c)

# A tuple can be instantiated without any elements
d = ()
print('d: ', d)
````

## Dictionary

````python
# A dictionary is a collection of key:value pairs

# A dictionary can be instantiated without any elements
sample_dict = {}

# Keys and Values support many data types and data structures
sample_dict['record_one'] = 1
sample_dict['The value of x is'] = x
sample_dict[2] = 'record_two'
sample_dict['fourth_entry'] = c

print(sample_dict)

# Dictionaries are often used when time and space constraints are an issue. It is HIGHLY performant.
````

````python
# Use type() to find out what type of object it is
print(type(sample_dict))
print(type(d))
print(type(x))
````
